### Permuted Featured Importance

#### Train permuted Random Forest

```python
## Split data into numeric and categorial categories
cols_to_drop = ["Interlayered Smectite","Microbiome", "Group"]
soil_scaled = soil_scaled.drop(cols_to_drop, axis=1)


## Save Response
soil_scaled['NTM Culture'].replace(('Positive', 'Negative'), (1,0), inplace=True)
response = list(soil_scaled['NTM Culture'])
soil_scaled3 = soil_scaled.drop("NTM Culture", axis=1)
```

```python
#### Variation among permutations
rf = RandomForestClassifier(random_state=1) 

def r2(rf, X_train, y_train):
    return r2_score(y_train, rf.predict(X_train))

perm_imp_rfpimp = permutation_importances(rf, soil_scaled3, response, r2)
```

```python
random.seed(1)
perm = PermutationImportance(rf, cv = "prefit", refit = False, n_iter = 1000).fit(soil_scaled3, response, seed=1)

x = list(perm.feature_importances_std_)
y = list(perm.feature_importances_)
z = list(soil_scaled3.columns)

df_importance = pd.DataFrame({'Names':z, 'Gini_Index':y, 'Average Variation':x})
df_importance['Gini_Index'] = pd.to_numeric(df_importance['Gini_Index'])
df_importance['Average Variation'] = pd.to_numeric(df_importance['Average Variation'])
df_importances = df_importance

importances = perm.feature_importances_
errors = perm.feature_importances_std_
indices = np.argsort(importances)[::-1]
df_importance = df_importance.transpose()
df_importance.columns = df_importance.iloc[0]
df_importance = df_importance[1:]
my_list = []
for i in indices:
    val = str(df_importance.columns[i])
    my_list.append(val)
    
```

#### Shapiro Wilks & Non-Parameteric Testing

```python
from scipy.stats import mannwhitneyu, shapiro, ttest_ind

High = soil_scaled[soil_scaled['NTM Culture']==1]
Low = soil_scaled[soil_scaled['NTM Culture']==0]

High = High.drop("NTM Culture", axis=1)
Low = Low.drop("NTM Culture", axis=1)

high_list = []
low_list = []

my_columns = []

## Test if sample distribution per feature are normalily distributed
for i in range(len(High.columns)):
    High2 = shapiro(np.array(High[High.columns[i]]))
    Low2 = shapiro(np.array(Low[Low.columns[i]]))
    my_columns.append(Low.columns[i])
    high_list.append(High2[1])
    low_list.append(Low2[1])

## Round to 2 significant figures    
high_list = ['{:g}'.format(float('{:.2g}'.format(i))) for i in high_list]
low_list = ['{:g}'.format(float('{:.2g}'.format(i))) for i in low_list]

## Combine Values to Matrix 
Normalilty_Matrix = pd.DataFrame(my_columns)
Normalilty_Matrix['Culture Positive'] = high_list
Normalilty_Matrix['Culture Negative'] = low_list
Normalilty_Matrix.set_index(0)
```

```python
Normal_Columns = ['1:1 Clay', 'Hematite']

## Subset Data Frames into Normal and NonNormal Columns
positive_norm = High[Normal_Columns]
positive_non_norm = High.drop(Normal_Columns, 1)

negative_norm = Low[Normal_Columns]
negative_non_norm = Low.drop(Normal_Columns, 1)

p_value_list = []

my_columns = []

## Run Nonparameteric Tests (Unequal Samples ## Need more to run)
for i in range(len(positive_non_norm.columns)):
    test = mannwhitneyu(np.array(positive_non_norm[positive_non_norm.columns[i]]), np.array(negative_non_norm[negative_non_norm.columns[i]]))
    my_columns.append(negative_non_norm.columns[i])
    p_value_list.append(test.pvalue)
    
    
## Run Parametric Tests
for i in range(len(positive_norm.columns)):
    test = ttest_ind(np.array(positive_norm[positive_norm.columns[i]]), np.array(negative_norm[negative_norm.columns[i]]), equal_var=False)
    my_columns.append(negative_norm.columns[i])
    p_value_list.append(test.pvalue)

    
## Combine Values into Matrix 
Significance_Matrix = pd.DataFrame(my_columns)
Significance_Matrix['P_Value'] = p_value_list
Significance_Matrix.columns = ['Feature', 'P_Value']


soil_scaled['NTM Culture'].replace((1, 0), ('Positive Culture', 'Negative Culture'), inplace=True)
Significance_Matrix
```
